2) Modifier l'image pour que tous les pixels soient blancs.<br>

```python
#Réponses:
print ("Question 1")
from PIL import Image

pic=Image.new('RGB',(130,130))  #L'image vierge a pour dimensions 130*130 car 16 900 = 130*130
pic.show()
pic.save('Noir.jpg')

print ("Question 2")
for x in range(130):
    for y in range(130):
        pic.putpixel((x,y),(255,255,255))

pic.show()
print("Blanc sur blanc, c'est normal que l'on ne le voie pas.")
pic.save('Blanc.jpg')
```

```python
#Correction

#1
print ("Question 1")
from PIL import Image

carré=Image.new('RGB',(130,130))  #L'image vierge a pour dimensions 130*130 car 16 900 = 130*130
carré.show()
carré.save('Carré.jpg')            
```

```python
#Correction

#2
print ("Question 2")
x =0
y=0
for loop in range(130):
    for loop in range(130):
        carré.putpixel((x,y),(255,255,255))
        x=x+1
    y = y+1
    x=0
carré.show()
print("Blanc sur blanc, c'est normal que l'on ne le voie pas.")
carré.save('carréblanc.jpg')
```

```python
#Réponse:

from PIL import Image
import random
monimg=Image.new('RGB',(50,50))
largeur,hauteur=monimg.size
for x in range(1,largeur-1):
    for y in range(1,hauteur-1):
        monimg.putpixel((x,y),(random.randint(0,255),random.randint(0,255),random.randint(0,255)))
monimg.show()
monimg.save('mon_image.jpg')

monimg2=Image.new('RGB',(50,50))
largeur,hauteur=monimg2.size
for x in range(1,largeur-1):
    for y in range(1,hauteur-1):
        monimg2.putpixel((x,y),(255-x*5,255-y*5,(x+y)//2*5))
monimg2.show()
monimg2.save('mon_image2.jpg')
```

```python
#Réponses:
#Aide utilisation des conditionnelles: if (si en anglais):

# if x<50 or 100<x<200:      --> Si x<50 OU x strictement compris entre 100 et 200 alors ...
#        action si c'est vrai
# else:
#      action si c'est faux
damierVB=Image.open('Blanc.jpg')
largeur, hauteur =damierVB.size
for x in range(0,largeur):
    for y in range(0,hauteur):
        if (x%2==0 and y%2==0) or (x%2==1 and y%2==1):
            damierVB.putpixel((x,y),(0,255,0))
damierVB.show()
damierVB.save('damierVertBlanc.png')
```
