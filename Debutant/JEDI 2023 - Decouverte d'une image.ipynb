```python
# question 1:
from PIL import Image

# question 2:
damier = Image.open('DamierRB.bmp')

# affichage de la largeur et hauteur
largeur,hauteur = damier.size
print("Largeur en pixels : ",largeur," Hauteur en pixels : ",hauteur)

# affichage de l'image
damier.show()
print("Cette image est composée de", largeur * hauteur, "pixels.")

# téléchargement de l'image pour l'ouvrir et zoomer
damier.save("damier.png")
```
