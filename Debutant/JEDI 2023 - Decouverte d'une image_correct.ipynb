```python
# observer l'image obtenue avec la fonction putpixel ci-dessous
damier = Image.open('DamierRB.bmp')
damier.putpixel((1,2),(0,0,255))
damier.show()
damier.save("damier_1.png")
```

```python
# compléter les pointillés pour modifier l'image avec des couleurs différentes
damier = Image.open('DamierRB.bmp')
damier.show()
largeur,hauteur = damier.size
for i in range (0,largeur):
    for j in range(0,hauteur):
        if(i%2==0 and j%2==0):
            damier.putpixel((i,j),(255,255,255))
        if(i%2==0 and j%2==1):
            damier.putpixel((i,j),(255,0,0))
        if(i%2==1 and j%2==0):
            damier.putpixel((i,j),(255,0,0))
        if(i%2==1 and j%2==1):
            damier.putpixel((i,j),(255,255,255))
damier.show()
damier.save("damier_2.png")
```
