# Activité 1 : Manipulation de fichiers texte et application aux fichiers CSV

L'informatique est avant tout la science du traitement de l'information. Une question centrale est donc de savoir comment représenter et manipuler cette information. En pratique, les données de toute nature (numérique, textuelles, images, audio...) sont représentées sous forme binaire puis enregistrer sur une mémoire persistente (comme un disque dur par exemple). Toutefois, manipuler au quotidient des données codées en binaire dans la mémoire n'est pas simple pour un humain...

Une abstraction courante est donc d'utiliser des **fichiers** qui représentent un découpage logique de la mémoire. Chaque fichier est associé à un **format**, définissant la manière dont le fichier sera représenté en mémoire.

Dans cette activité, nous nous concentrerons sur la manipulation de **fichiers texte**. Nous verrons comment **lire** un fichier et récupérer les informations qu'il contient mais également comment **écrire** dans un fichier.

## Fichier texte brut

Les fichiers au format texte brut n'ont pas de structure particulière et contiennent simplement une suite de caractères. Le nom de ces fichiers se termine par `.txt` (plus rarement `.text`).

La fonction suivante permet de créer un fichier texte.

```python
def ecriture(nomFichier:str) -> None:
    fichier = open(nomFichier+".txt", "w",encoding="utf-8") 
    fichier.write("ceci est la ligne 1 \n") 
    fichier.write("ceci est la ligne 2 \nceci est la ligne 3\n")
    fichier.write("ceci est la ligne 4 ")
    fichier.write("ceci est toujours la ligne 4\n")
    a=input("Saisir un texte à écrire dans le fichier : ")
    fichier.write("Message saisi par l'utilisateur : "+a+"\n")
    fichier.close()
    print("Le fichier ",nomFichier,".txt a été créé")
    
ecriture("fichierTest")
```

!!! note Détails sur le code

La fonction `open()` créer un **objet fichier** (aussi appelé *descripteur de fichier* dans d'autres langages). Cet objet peut être vu comme un pointeur vers le fichier permettant de le manipuler.

Le premier paramètre indique le **nom du fichier** à manipuler.

Le deuxième paramètre décrit le **mode** dans lequel le fichier sera ouvert. Les différents modes sont les suivants:

- Le caractère `'w'` (pour *write* en anglais) correspond au mode **écriture**. Si le fichier n'existe pas, il sera créé. Attention! S'il existe déjà, il sera écrasé! 
- Le caracatère `'a'` (pour *append*) correspond au mode **ajout** et permet d'ajouter des lignes à un fichier existant.
- Le caractère `'r'` correspond au mode **lecture seule**.
- Le caractère `'r+'` correspond au mode **lecture/écriture**.

Les instructions de la forme `fichier.write(mon_texte)` vont rajouter le texte passé en paramètre à la fin du fichier. Le caractère `\n` indique un retour à la ligne.

Pour finir, il faut penser à **fermer le fichier** avec l'instruction `fichier.close()`. L'utilisation du mot-clé `with` permet d'éviter d'avoir à prendre cette précaution, mais nous verrons plus tard comment l'utiliser.

Pour aller plus loin, c'est par [içi](https://docs.python.org/3/tutorial/inputoutput.html#reading-and-writing-files).
!!!

<div class = "alert alert-info"> 
    
Contrairement à l'instruction `print()` qui accepte plusieurs argurments, par exemple `print("valeur de a :",a)`, `fichier.write()` ne peut accepter qu'un seul argument et celui-ci doit être une chaîne de caractères.
    
Ainsi, pour écrire un entier, il faut le convertir avec l'instruction `str()`.

```python
def lecture(nomFichier: str) -> None:
    fichier = open(nomFichier, "r",encoding="utf-8")  # r pour read
    ligne=fichier.readline()
    while (ligne!=""):    # boucle while : tant que la ligne n'est pas vide, on execute les 3 lignes suivantes
        ligne=ligne.strip()   # permet de se débarasser d'un retour à la ligne éventuel \n
        print(ligne)
        ligne=fichier.readline()
    fichier.close()

lecture("fichierTest.txt")
```

## Application au format CSV

Un autre format de données très utilisé sur le web est le format **CSV** pour **C**omma **S**eparated **V**alues. La cellule suivante fournit une brève description de ce format.

!!! info  Présentation du format CSV

**L'encodage**

Un fichier CSV est un fichier texte (c'est-à-dire que les octets qui le composent doivent être interprétés comme des codes de caractères).

Comme presque toujours sur le Web, les caractères sont bien encodés en **UTF-8** (il convient de garder l'encodage proposé par défaut dans notre cas).

**Séparateur de cellules**

Les données contenues dans un fichier CSV correspondent à des **données tabulées**. Les lignes sont naturellement séparées par un caractère de saut de ligne (celui qui correspond à la touche Entrée du clavier).

Les cellules d'une même ligne sont en théorie séparées par une **virgule** (*comma* signifie d'ailleurs *virgule* en anglais). Toutefois il est courant pour les fichiers en langue française de privilégier l'usage du **point virgule**. Cette précaution est dû au fait que les valeurs numériques flottantes utilisent une virgule pour séparer la partie entière de la partie décimale, menant à une confusion parfois facheuse...

**Enregistrements et descripteurs**

Chaque ligne du fichier correspond à un **enregistrement** dans la table. Chaque colonne du fichier correspond à un attribut de l'enregistrement, décrit par un **descripteur**. Bien souvent la première ligne d'un fichier csv fournit l'ensemble des descripteurs du fichier. 
!!!

<div class = "alert alert-info">
    
S'il est facile de créer un fichier au format `text`, il est plus délicat d'en extraire des données, car ce type de fichier n'est pas structuré.
    
Aussi, il est préférable de manipuler des fichiers au format `csv` qui correspond à une feuille de calcul d'un tableur.
    
On trouve de nombreux fichiers de données sur internet sous format `csv`.

### Lecture et écriture sans module

Supposons que nous disposons d'un relevé de note sous la forme d'une liste de listes Python appelée `notes`. 
Afin d'écrire un fichier `csv` à partir de `notes`, nous allons procéder de manière très similaire à l'écriture d'un fichier texte brute.

```python
def ecriture_csv(nom_fichier: str,datas: list[list[str]],separateur=",") -> None:
    with open(nom_fichier+".csv","w",encoding="utf-8") as fichier:
        for i in range(len(datas)):
            ligne = ""
            for j in range(len(datas[i])-1):
                ligne += str(datas[i][j])+separateur
            ligne += str(datas[i][-1]) + "\n"   # en Python, l'indice -1 permet d'accéder directement au dernier élément" au début de lecture/écriture de fichier csv
            fichier.write(ligne)
            
ecriture_csv("test",[[1,2,3],[4,5,6]])            
```

Si vous exécutez ce code depuis Capytale, le fichier `test.csv` ainsi généré est accessible en cliquant sur le bouton "liste des fichiers annexes (le trombone)/Disponible le temps de la session/text.csv".

!!! note Détails sur le code
Le paramètre `encoding` de la fonction `open()` permet de préciser que l'encodage du fichier se fera en `utf-8`.

L'instruction `with` permet de créer un fichier objet et de gérer la fermeture du fichier à la fin du bloc.

Pour chaque élément de `datas`, on crée une chaine de caractère nommé `ligne` où les valeurs sont séparées par une virgule. Ainsi une liste `[1,2,3]` sera transformé en `"1,2,3\n"`. On remarquera à nouveau l'utilisation du caractère `\n` indiquant un retour à la ligne. Ce caractère est bien sur invisible et n'apparaitra pas à l'affichage.

Une fois la ligne créée, on l'ajoute au fichier avec l'instruction `fichier.write(ligne)`.

## Un avant-goût de l'après-midi : représentation graphique de données

Il est possible de visualiser les données grâce au module [matplotlib](https://matplotlib.org/). Nous présenterons plus en détails ce module cet après-midi mais voici déjà une illustation simple de l'utilisation de ce module.

```python
# Import de la bibliothèque, utilisation de l'alias plt
import matplotlib.pyplot as plt
# tableau de couleurs qui seront utilisées pour le diagramme
colors=["lightcoral","fuchsia","cyan","pink","moccasin","lime","lightblue","plum","bisque"]
  
# les valeurs en abscisse    
x=[0,8,12,13,20]
# les valeurs en ordonnées
resultat=[100,20,45,90,60]
# les labels de chaque barre
nom=['A','B','C','D','E']

# On dessine le diagramme à partir des données fournies
plt.bar(x,resultat,width = 0.6 ,color=colors) 
# On ajoute le label des barres
plt.xticks(x, nom,  rotation = 80)
# on ajoute un titre au diagramme
plt.title("Répartition ")
# on affiche le tout
plt.show()
plt.close()
```
